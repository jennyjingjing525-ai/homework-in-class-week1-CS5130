5. Create a VectorVisualizer object and plot the vectors $\vec{a} = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$, $\vec{b} = \begin{pmatrix} 3 \\ 0 \end{pmatrix}$, and $\vec{c} = \begin{pmatrix} 0 \\ 3 \end{pmatrix}$ on the same graph. What shape do these vectors form?

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Define the vectors
vectors = [
    {"coords": [1, 1], "label": "Vector 1", "color": "blue"},
    {"coords": [3, 0], "label": "Vector 2", "color": "green"},
    {"coords": [0, 3], "label": "Vector 3", "color": "red"}
]

# Plot each vector from origin
for vec in vectors:
    ax.quiver(0, 0, vec["coords"][0], vec["coords"][1], 
              angles='xy', scale_units='xy', scale=1,
              color=vec["color"], label=vec["label"], width=0.015)

# Extract the endpoints of the vectors
points = [vec["coords"] for vec in vectors]
points.append(points[0])  # Close the triangle

# Convert to numpy array for easier manipulation
points = np.array(points)

# Plot the triangle formed by connecting the endpoints
ax.plot(points[:, 0], points[:, 1], 'o--', color='purple', linewidth=2, markersize=8)

# Add labels for the points
for i, point in enumerate(points[:-1]):  # Skip the last point (duplicate of first)
    ax.text(point[0]+0.1, point[1]+0.1, f'P{i+1} ({point[0]}, {point[1]})', fontsize=12)

# Set the limits of the plot
ax.set_xlim(-0.5, 4)
ax.set_ylim(-0.5, 4)

# Add grid and labels
ax.grid(True, linestyle='--', alpha=0.7)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Vectors and the Triangle Formed by Their Endpoints')

# Add legend
ax.legend(loc='upper right')

# Set aspect ratio to equal
ax.set_aspect('equal')

# Display the plot
plt.show()
```

8. Calculate the dot product of $\vec{a} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}$ and $\vec{b} = \begin{pmatrix} 2 \\ 2 \end{pmatrix}$. Then calculate the magnitudes of both vectors and the angle between them using the formula $\vec{a} \cdot \vec{b} = |\vec{a}||\vec{b}|\cos\theta$.

```python
import numpy as np
vector_a=np.array([1,2])
vector_b=np.array([2,2])
dot_product = dot_product(vector_a, vector_b)
magnitude_a = np.linalg.norm(vector_a)
magnitude_b = np.linalg.norm(vector_b)
print(magnitude_a)
print(magnitude_b)
# Calculate the cosine of the angle
cos_theta = dot_product / (magnitude_a * magnitude_b)

# Calculate the angle in radians and convert to degrees
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print(f"a · b = {dot_product}")
print(f"||a|| = {magnitude_a:.4f}")
print(f"||b|| = {magnitude_b:.4f}")
print(f"cos(θ) = {cos_theta:.4f}")
print(f"Angle between vectors = {theta_rad:.4f} radians or {theta_deg:.2f} degrees")
```

12. Create a 2×2 matrix that rotates vectors by 60 degrees counterclockwise. Apply this matrix to the vector $\vec{v} = \begin{pmatrix} 2 \\ 0 \end{pmatrix}$ and verify the result is correct.

```python
# Rotation matrix (60 degrees)
import math
import numpy as np
theta = math.pi/3
R = [[math.cos(theta), -math.sin(theta)], 
      [math.sin(theta), math.cos(theta)]]
rotation_matrix=np.array(R)
print("Rotation matrix:\n", rotation_matrix)

# define vector to rotate
vector=np.array([2,0])

# Apply the rotation
rotated_vector = np.dot(rotation_matrix, vector)

print(f"\nOriginal vector: {vector}")
print(f"Rotated vector: {rotated_vector}")
```

15. Verify that $\det(AB) = \det(A) \cdot \det(B)$ for the matrices $A = \begin{pmatrix} 2 & 1 \\ 3 & 2 \end{pmatrix}$ and $B = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}$.

```python
import numpy as np
A = np.array([[2, 1], [3, 2]])
det_A = np.linalg.det(A)
B =np.array([[1,2],[3,4]])
det_B = np.linalg.det(B)
C = A@B
D = det_A*det_B
det_C = np.linalg.det(C)
print(f"Determinant of A: {det_A}")
print(f"Determinant of B: {det_B}")
print(f"Determinant of AB: {det_C}")
print(D)
print(f"\n验证 det(AB) = det(A)det(B): {np.isclose(det_C, D)}")
```

16. Create a 2×2 matrix with a determinant of zero. What happens when you try to find its inverse? Visualize this result to verify.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a singular matrix (determinant = 0)
A = np.array([[1, 2], 
              [2, 4]])  # Second row is 2 times the first row

print("Matrix A:")
print(A)

# Calculate determinant
det_A = np.linalg.det(A)
print(f"\nDeterminant of A: {det_A}")

# try inverse
try:
    A_inv = np.linalg.inv(A)
    print("A inverse:\n", A_inv)
    print("\nA × A_inv:\n", np.dot(np.array(A), A_inv))
except np.linalg.LinAlgError as e:
    print(f"\nError when trying to compute inverse: {e}")
    print("This is expected because the matrix is singular (determinant = 0).")
```

17. Calculate the inverse of $A = \begin{pmatrix} 3 & 1 \\ 2 & 1 \end{pmatrix}$. Verify that $A \times A^{-1} = I$.

```python
import numpy as np

A = np.array([[3, 1], 
              [2, 1]])  
A_inv = np.linalg.inv(A)
print (A)
print(A_inv)
# Multiply A by its inverse
product = np.dot(A, A_inv)
print("\nA × A⁻¹:")
print(product)

# Check if the result is approximately the identity matrix
identity = np.eye(2)  # 2x2 identity matrix
is_identity = np.allclose(product, identity)
print(f"\nIs A × A⁻¹ approximately equal to the identity matrix? {is_identity}")

# Show the difference (should be very small)
difference = product - identity
print("\nDifference from identity matrix:")
print(difference)
```
